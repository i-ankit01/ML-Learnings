```python
df = pd.read_csv('ipl-matches.csv')
df.sample(3)
```

```python
def rename(name):
    result = ''
    words = name.split()
    for word in words:
        result += word[0]
    return result
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('ipl-matches.csv', converters={'team1' : rename , 'team2' : rename, 'toss_winner' : rename} , usecols=['id', 'city', 'team1', 'team2', 'toss_winner'])
```
