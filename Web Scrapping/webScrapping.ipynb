# Web Scrapping


```python
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Web_page'
webpage = requests.get(url)
```

```python
webpage # If response is 403 it means that the website doesn't allow bots so we will add a header and then send the request
```

```python
# headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
content = requests.get(url).text
content
```

```python
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
subheadings = []
for i in soup.find_all('h2'):
    subheadings.append(i.text.strip())
subheadings
```

```python
first_paragraph = soup.find_all('p')[1].text.strip()
first_paragraph
```

```python
reference_list = soup.find('div', class_="reflist").text.strip()
reference_list
```
